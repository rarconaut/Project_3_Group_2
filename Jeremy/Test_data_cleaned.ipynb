```python
import pandas as pd
import os
from sqlalchemy import create_engine
from config import my_password
```

```python
test_df = pd.read_csv('aug_test.csv')
```

```python
test_df.head()
```

```python
test_df.drop(['city_development_index', 'last_new_job', 'training_hours'], axis = 'columns', inplace = True)
test_df.head()
```

```python
test_df.info()
```

```python
test_df.dropna(how='any', inplace=True)
test_df.info()
```

```python
test_df.to_csv('test_cleaned.csv', index=False)
```

```python
male_filter = test_df['gender'] == 'Male'
test_male_df = test_df[male_filter]
test_male_df.head()
```

```python
female_filter = test_df['gender'] == 'Female'
test_female_df = test_df[female_filter]
test_female_df.head()
```

```python
test_male_df.info()
```

```python
test_female_df.info()
```

```python
male_sample_df = test_male_df.sample(n=85)
male_sample_df.head()
```

```python
merged_df = pd.concat([male_sample_df, test_female_df], axis=0)
merged_df.head()
```

```python
merged_df.info()
```

```python
merged_df.to_csv('test_balanced.csv', index=False)
```
